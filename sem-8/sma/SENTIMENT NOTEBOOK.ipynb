```python
GB_comments = pd.read_csv("GBCommentsclean.csv", error_bad_lines=False) #Change in the code
```

```python
GB_comments.head()
```

```python
GB_comments['comment_text'] = GB_comments['comment_text'].str.replace("[^a-zA-Z#]", " ")
```

```python
GB_comments['comment_text'] = GB_comments['comment_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
GB_comments['comment_text'] = GB_comments['comment_text'].apply(lambda x:x.lower())
```

```python
tokenized_tweet = GB_comments['comment_text'].apply(lambda x: x.split())
tokenized_tweet.head()
```

```python
GB_comments = pd.read_csv("GBCommentssentiments.csv", error_bad_lines=False) #Change in the code
```

```python
GB_comments['Sentiment'] = GB_comments['Sentiment Scores'].apply(lambda s : 'Positive' if s > 0 else ('Neutral' if s == 0 else 'Negative'))
```

```python
GB_comments.head()
```

```python
GB_comments.Sentiment.value_counts()
```

```python
all_words = ' '.join([str(text) for text in GB_comments['comment_text']])
```

```python
import matplotlib.pyplot as plt 
from wordcloud import WordCloud
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words)
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
all_words_posi = ' '.join([text for text in GB_comments['comment_text'][GB_comments.Sentiment == 'Positive']])
```

```python
wordcloud_posi = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words_posi)      
plt.figure(figsize=(10, 7))                    
plt.imshow(wordcloud_posi, interpolation="bilinear")                     
plt.axis('off')             
plt.show()
```

```python
all_words_nega = ' '.join([text for text in GB_comments['comment_text'][GB_comments.Sentiment == 'Negative']])
```

```python
wordcloud_nega = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words_nega)      
plt.figure(figsize=(10, 7))                    
plt.imshow(wordcloud_nega, interpolation="bilinear")                     
plt.axis('off')             
plt.show()
```
