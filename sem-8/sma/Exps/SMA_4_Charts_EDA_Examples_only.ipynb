```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
!pip install squarify
import squarify
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate sample data
data = np.random.randint(1, 11, size=50)
```

```python
# 1. Pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
sizes = np.random.randint(1, 11, size=len(labels))

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')
plt.title('Pie Chart')
plt.show()
```

```python
# 2. Area chart
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.fill_between(x, y)
plt.title('Area Chart')
plt.show()
```

```python
# 3. Violin plot
data = np.random.normal(size=100)

sns.violinplot(data=data)
plt.title('Violin Plot')
plt.show()
```

```python
# 4. Bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 11, size=len(categories))

plt.bar(categories, values)
plt.title('Bar Chart')
plt.show()
```

```python
# 5. Line chart
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.title('Line Chart')
plt.show()
```

```python
# 6. Scatter plot
x = np.random.normal(size=50)
y = np.random.normal(size=50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.show()
```

```python
# 7. Heatmap
data = np.random.randint(1, 11, size=(5, 5))

sns.heatmap(data, annot=True, cmap='Blues')
plt.title('Heatmap')
plt.show()
```

```python
# 8. Box plot
data = np.random.normal(size=100)

sns.boxplot(data=data)
plt.title('Box Plot')
plt.show()
```

```python
# 9. Treemap
values = np.random.randint(1, 11, size=8)
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E', 'Category F', 'Category G', 'Category H']
colors = sns.color_palette('Paired', n_colors=len(labels))

fig, ax = plt.subplots(figsize=(8, 8))
squarify.plot(sizes=values, label=labels, color=colors, ax=ax)
plt.axis('off')
plt.title('Treemap')
plt.show()
```

```python
# 10. Bubble chart
x = np.random.normal(size=50)
y = np.random.normal(size=50)
z = np.random.randint(1, 11, size=50)

plt.scatter(x, y, s=z*100)
plt.title('Bubble Chart')
plt.show()
```
